```python
import pandas as pd
import matplotlib
import numpy as np #importing the tools I need 

df_csv3 = pd.read_csv("data_all_regions_raw.csv") #opening up my file 
df_csv3 #turning my file into a Dataframe that I'll make changes to 
```

```python
df_csv3['Order_Date'] = df_csv3['Order_Date'].str.replace('-', '/') #cleaned the order and ship date columns to match the format
```

```python
# Ensure Order_Date is a datetime object
df_csv3['Order_Date'] = pd.to_datetime(df_csv3['Order_Date'])

# Convert Order_Date to the desired format (MM-DD-YYYY)
df_csv3['Order_Date'] = df_csv3['Order_Date'].dt.strftime('%m-%d-%Y')
```

```python
# Replace '-' with '/' in the Ship_Date column
df_csv3['Ship_Date'] = df_csv3['Ship_Date'].str.replace('-', '/')
```

```python
# Ensure Ship_Date is a datetime object after the replacement
df_csv3['Ship_Date'] = pd.to_datetime(df_csv3['Ship_Date'])

# Convert Ship_Date to the desired format (MM/DD/YYYY)
df_csv3['Ship_Date'] = df_csv3['Ship_Date'].dt.strftime('%m/%d/%Y')
```

```python
# Ensure Order_Date is a datetime object
df_csv3['Order_Date'] = pd.to_datetime(df_csv3['Order_Date'])

# Split the date into separate columns for year, month, and day
df_csv3['Year'] = df_csv3['Order_Date'].dt.year
df_csv3['Month'] = df_csv3['Order_Date'].dt.month
df_csv3['Day'] = df_csv3['Order_Date'].dt.day
```

```python
# Ensure Ship_Date is a datetime object
df_csv3['Ship_Date'] = pd.to_datetime(df_csv3['Ship_Date'])

# Populate the Ship_Year, Ship_Month, and Ship_Day columns
df_csv3['Ship_Year'] = df_csv3['Ship_Date'].dt.year
df_csv3['Ship_Month'] = df_csv3['Ship_Date'].dt.month
df_csv3['Ship_Day'] = df_csv3['Ship_Date'].dt.day
```

```python
# Lowercase all column names
df_csv3.columns = df_csv3.columns.str.lower()
```

```python
# Drop the 'unnamed: 0' column
df_csv3 = df_csv3.drop(columns=['unnamed: 0'])
```
